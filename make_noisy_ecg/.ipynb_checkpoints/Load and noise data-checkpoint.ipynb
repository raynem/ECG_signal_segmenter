```python
import numpy as np
import matplotlib.pylab as plt
import scipy.signal
import wfdb
import os
```

```python
data = open('emg_healthy.txt', encoding = "ISO-8859-1").readlines()
```

```python
timeseries = []
for dd in data:
    try:
        d = float(dd.split(' ')[1])
    except:
        continue
    timeseries.append(d)
timeseries_decimate = scipy.signal.decimate(timeseries, 8)
```

```python
timeseries_decimate.shape
```

```python
files = ['./xxx/Person_' + str(i).zfill(2) + '/rec_1' for i in range(1, 90)]
```

```python
recs = [wfdb.rdrecord(f) for f in files]
```

```python
plt.figure()
plt.plot(timeseries_decimate[:1000])
plt.show()
```

```python
noise = timeseries_decimate
X, Y = [], []

for rec in recs:
    signal = rec.p_signal[:, 1]
    for i in range(0, len(signal), 100): # record loop
        for j in range(0, len(noise), 100): # noise loop
            try:
                noise_ts = noise[i:i+1000]
                ecg_ts = signal[i:i+1000]
                merged_ts = noise_ts + ecg_ts
            except:
                continue
            X.append(merged_ts)
            Y.append(ecg_ts)
```

```python
%matplotlib inline
```

```python
plt.figure(figsize = (15, 7))
plt.plot(X[1])
plt.plot(Y[1])
plt.show()
```
